```python
import numpy as np
from sklearn.metrics import accuracy_score
from ml_from_scratch.tree import DecisionTreeClassifier
import pandas as pd
```

```python
from sklearn.datasets import load_digits
from sklearn.model_selection import train_test_split

# LOAD DATA
digits = load_digits()
X = digits.data
y = digits.target
```

```python
# SPLIT DATA

X_train, X_test, y_train, y_test = train_test_split(X,
                                                    y,
                                                    test_size=0.3,
                                                    stratify=y,
                                                    random_state=3)
```

```python
# CLASSIFICATION - A Very Fit Tree
# Create a decision tree classification
# Using hyperparameter custom criteria with default max depth
clf_gini = DecisionTreeClassifier(criteria='gini') # Menggunakan Gini
clf_entropy = DecisionTreeClassifier(criteria='entropy') # Menggunakan Entropy
clf_log_loss = DecisionTreeClassifier(criteria='log_loss') # Menggunakan Entropy
clf_gini.fit(X_train, y_train)
clf_entropy.fit(X_train, y_train)
clf_log_loss.fit(X_train, y_train)
```

```python
# Predict & Calculate Accuracy score test
y_pred_train_gini = clf_gini.predict(X_train)
y_pred_test_gini = clf_gini.predict(X_test)

y_pred_train_entropy = clf_entropy.predict(X_train)
y_pred_test_entropy = clf_entropy.predict(X_test)

y_pred_train_log_loss = clf_log_loss.predict(X_train)
y_pred_test_log_loss = clf_log_loss.predict(X_test)

print(f"Accuracy train using Gini : {round(accuracy_score(y_train, y_pred_train_gini),5)}")
print(f"Accuracy test using Gini : {round(accuracy_score(y_test, y_pred_test_gini),5)}")
print("")
print("")
print(f"Accuracy train using Entropy : {round(accuracy_score(y_train, y_pred_train_entropy),5)}")
print(f"Accuracy test using Entropy : {round(accuracy_score(y_test, y_pred_test_entropy),5)}")
print("")
print("")
print(f"Accuracy train using Log Loss : {round(accuracy_score(y_train, y_pred_train_log_loss),5)}")
print(f"Accuracy test using Log Loss : {round(accuracy_score(y_test, y_pred_test_log_loss),5)}")
```

```python
# Define parameter possibilities as lists
p_criterion = ['gini', 'entropy', 'log_loss']
p_max_depth = [1,3,5,7,9,12]
# Define place holder for score
results = []

# Nested loops - we need to test for all combinations
for criterion in p_criterion:
    for max_depth in p_max_depth:
        # Train the model
        model = DecisionTreeClassifier(
            criteria=criterion,
            max_depth=max_depth
        )
        model.fit(X_train, y_train)
        trains = model.predict(X_train)
        preds = model.predict(X_test)
        # Append current results
        results.append({
            'Accuracy train': round(accuracy_score(y_train, trains), 5),
            'Accuracy test': round(accuracy_score(y_test, preds), 5),
            'P_Criterion': criterion,
            'P_MaxDepth': max_depth
        })
        
results = pd.DataFrame(results)
results = results.sort_values(by='Accuracy test', ascending=False)
results
```
